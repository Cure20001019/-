```python
# %load 1-resize.py
# To add a new cell, type '# %%'
# To add a new markdown cell, type '# %% [markdown]'

# %%
import os
import cv2
import numpy as np


# 读取图像，解决imread不能读取中文路径的问题
def cv_imread(filePath):
    # 核心就是下面这句，一般直接用这句就行，直接把图片转为mat数据
    cv_img=cv2.imdecode(np.fromfile(filePath,dtype=np.uint8),-1)
    # imdecode读取的是rgb，如果后续需要opencv处理的话，需要转换成bgr，转换后图片颜色会变化
    cv_img=cv2.cvtColor(cv_img,cv2.IMREAD_COLOR)
    return cv_img

def cv_write(filePath,img):
    cv2.imencode('.jpg',img)[1].tofile(filePath)

# # path = r'C:\Users\air10\Desktop\IMG_1844.jpg'
# path = r'D:\ZxH\Active Files\地标识别实训课\新收集数据\1_第一教学楼_E\IMG_1657.jpg'
# # Using cv2.imread() method
# img = cv_imread(path)
# img = cv2.resize(img,None,fx=0.2,fy=0.2)
# print(img.shape)
# # Displaying the image
# cv2.imshow('image', img)
# cv2.waitKey()
# cv2.destroyAllWindows()

# %%
def batch_resize(src_dir, dst_dir, img_size, landmark):
    for img in os.listdir(os.path.join(src_dir, landmark)):
        org_array = cv_imread(os.path.join(src_dir, landmark, img))
        new_array = cv2.resize(org_array, img_size, interpolation=cv2.INTER_AREA)
        cv_write(os.path.join(dst_dir, img), new_array)

    print("{} finished.".format(landmark))


# %%
if __name__ == '__main__':
    src_dir = r'C:\Users\air10\Desktop\New Data'
    dst_dir = src_dir + "_edited" 
    img_size=(512,512)
    i=0

    if not os.path.exists(dst_dir):
        os.makedirs(dst_dir)
        print("{} created.".format(dst_dir)) 

    for landmark in os.listdir(src_dir):
        if os.path.exists(os.path.join(dst_dir, landmark)):
            print("{} existed.".format(landmark))
            continue
        else:
            save_path = os.path.join(dst_dir, landmark)
            os.mkdir(save_path)
            batch_resize(src_dir, save_path, img_size, landmark)
            i+=1
    print("{:>3d} folders processed in total.".format(i))
```
